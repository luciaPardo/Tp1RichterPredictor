## Tipo de Daño segun Cantidad de Pisos

```python
daño_cant_pisos = values.loc[:,["building_id","count_floors_pre_eq"]]\
.merge(labels,on="building_id",how='inner')\
.loc[:,["damage_grade","count_floors_pre_eq"]]
#daño_cant_pisos = pd.pivot_table(daño_cant_pisos,index="count_floors_pre_eq",columns="damage_grade",aggfunc='count')
daño_cant_pisos['aux'] = 1
daño_cant_pisos = daño_cant_pisos.pivot_table(index='count_floors_pre_eq'\
                        ,columns='damage_grade',values='aux',aggfunc='count')

for i in range (len(daño_cant_pisos)):
    daño_cant_pisos.iloc[i] = daño_cant_pisos.iloc[i]/(daño_cant_pisos.iloc[i].sum())
```

# twilight_r,mako,blues,greens

```python
fig, ax = plt.subplots(figsize=(14,10)) 
p1 = sns.heatmap(daño_cant_pisos,annot=True,cmap='Greens')
#annot true por defecto pone el formato en cientifico
ax.set_ylabel('Cantidad de pisos',fontsize=14)
ax.set_xlabel('Tipo de daño',fontsize=14)
ax.set_title('Tipo de Daño segun Cantidad de Pisos',fontsize=18)

y_labels = daño_cant_pisos.index
x_labels = daño_cant_pisos.columns
aux = ax.set_yticklabels(y_labels,rotation=0,fontsize=12)
aux = ax.set_xticklabels(x_labels,fontsize=12)
fig.savefig("tipo_de_daño_segun_cant_pisos.png")
```

## Geo levels id 3 con promedio de daño igual a 3 por Cantidad de edificios

```python
geo_level_damage = values.loc[:,["geo_level_3_id", 'damage_grade']].groupby("geo_level_3_id").mean().reset_index()
geo_level_damage = geo_level_damage.loc[geo_level_damage.damage_grade==3]
bid = values.loc[:,["geo_level_3_id", 'building_id']].groupby("geo_level_3_id").count().reset_index()
geo_level_damage = geo_level_damage.merge(bid,on="geo_level_3_id",how='inner').nlargest(10,"building_id").rename(columns={"building_id":"amount_buildings"})
geo_level_damage
```

```python
#sería como los "geo lvl 3 id más afectados". Mientras más se sube el promedio menos cantidad de edificios hay.
f = geo_level_damage.loc[:,["geo_level_3_id","amount_buildings"]].set_index("geo_level_3_id")
ax = f.plot.bar(color='seagreen',rot=0,figsize=(10,10),fontsize=14)
ax.set_ylabel('Cantidad de contrucciones',fontsize=14)
ax.set_xlabel('Geo level id 3',fontsize=14)
ax.set_title('Geo levels id 3 con promedio de daño igual a 3 por Cantidad de edificios',fontsize=18)
ax.legend().remove()
```

# Columnas numericas

```python
train[['count_floors_pre_eq', 'age', 'area_percentage', 'height_percentage', 'count_families']]\
.describe()
```

```python
numericas = train[['damage_grade', 'count_floors_pre_eq', 'age', 'area_percentage', \
                   'height_percentage', 'count_families']]
numericas
```

```python
volumen = (numericas['height_percentage'] * numericas['area_percentage'])\
            .to_frame('volumen_percentage')
volumen.describe()
```

```python
#Poniendo una cota superior de 100 veo la mayoria de los datos

g = volumen.loc[volumen.volumen_percentage < 100]
g
```

```python
ax = sns.displot(g, kde=True, aspect=1.2, height=7, legend=False, palette='Greens')
axes = ax.axes.flatten()
axes[0].set_title('Distribucion de la cantidad de edificios por volumen normalizado',fontsize=18)
axes[0].set_xlabel('Volumen normalizado', fontsize=14)
axes[0].set_ylabel('Cantidad de edificios', fontsize=14)
#cellNumber+=1
#outPNG = 'visualizaciones/figure_{}'.format(cellNumber)
#plt.savefig(outPNG, bbox_inches='tight')
```

```python
numericas['volumen_percentage'] = volumen['volumen_percentage']
numericas
```

```python
volumen_damage = numericas.loc[numericas.volumen_percentage < 100, \
                               ['damage_grade', 'volumen_percentage']]
volumen_damage
```

```python
ax = sns.displot(volumen_damage, kind='kde', x='volumen_percentage', hue='damage_grade', \
           aspect=1.2, height=7, palette='Set2')
axes = ax.axes.flatten()
axes[0].set_title('Distribucion de la cantidad de edificios por volumen normalizado, segun tipo de daño',fontsize=18)
axes[0].set_xlabel('Volumen normalizado', fontsize=14)
axes[0].set_ylabel('Densidad', fontsize=14)
#cellNumber+=1
#outPNG = 'visualizaciones/figure_{}'.format(cellNumber)
#plt.savefig(outPNG, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax = sns.violinplot(data=volumen_damage, x='damage_grade', y='volumen_percentage', palette='Greens')
ax.set_title('Distribucion de la cantidad de edificios segun su volumen normalizado y tipo de daño', fontsize=18)
ax.set_xlabel('Tipo de daño', fontsize=14)
ax.set_ylabel('Volumen normalizado', fontsize=14)
#cellNumber+=1
#outPNG = 'visualizaciones/figure_{}'.format(cellNumber)
#plt.savefig(outPNG, bbox_inches='tight')
```

```python
#Discretizo los valores de familias en 'bins'
```

```python
bins = [0, 1, 2, 10]
labels = ['Ninguna', 'Una familia', 'Más de una familia']
familias = pd.cut(numericas['count_families'], bins, right=False, labels=labels)
familias.value_counts()
```

```python
numericas['familias'] = familias
g=numericas[['damage_grade', 'familias']]
g['hola'] = 1
g = g.pivot_table(index='damage_grade', columns='familias', values='hola', aggfunc='sum')
g
```

```python
fig, ax = plt.subplots(figsize=(14,10))
g=g.apply(lambda x: 100 * x / float(x.sum()))
ax = sns.heatmap(g,cmap='Greens',annot=True)
ax.set_xlabel('Cantidad de familias',fontsize=14)
x_labels = g.columns
aux = ax.set_xticklabels(x_labels,fontsize=12)
y_labels = g.index
aux = ax.set_yticklabels(y_labels, rotation=0, fontsize=12)
ax.set_ylabel('Tipo de daño',fontsize=14)
ax.set_title('Tipo de daño segun la cantidad de familias',fontsize=18)
for t in ax.texts: t.set_text(t.get_text() + " %")

#cellNumber+=1
#outPNG = 'visualizaciones/figure_{}'.format(cellNumber)
#plt.savefig(outPNG, bbox_inches='tight')
```

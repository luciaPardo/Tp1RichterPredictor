```python
import pandas as pd
import numpy as np
```

```python
lightGBM = pd.read_csv( 'Submits/LightGBM.csv')
XGBoost = pd.read_csv( 'Submits/XGBoostWithGSandFE.csv')
```

```python
XGBoost = XGBoost.rename(columns={'damage_grade': 'damage_grade_1'})
lightGBM = lightGBM.rename(columns={'damage_grade': 'damage_grade_2'})
```

```python
merge = XGBoost.merge(lightGBM)
damage_grade_1 = merge['damage_grade_1'].to_numpy()
damage_grade_2 = merge['damage_grade_2'].to_numpy()
```

```python
ensamble = []
for i in range (len(damage_grade_1)):
    if damage_grade_1[i] > damage_grade_2[i]:
        ensamble.append(damage_grade_1[i])
    else:
        ensamble.append(damage_grade_2[i])
ensamble
```

```python
merge['ensamble'] = ensamble
```

```python
merge
```

```python
my_submission = merge.set_index('building_id')
my_submission = my_submission.loc[:,['ensamble']]
my_submission = my_submission.rename(columns={'ensamble':'damage_grade'})
my_submission
```

```python
my_submission.to_csv('Submits/Ensamble.csv')
```
